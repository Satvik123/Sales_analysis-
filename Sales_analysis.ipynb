```python
files=[file for file in os.listdir(r'c:\Users\hp\Documents\Data analytics\Datasets\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data')]
```

```python
amd=pd.DataFrame()
files
```

```python
for file in files:
  df=pd.read_csv(r'c:\Users\hp\Documents\Data analytics\Datasets\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\\' + file)
  amd= pd.concat([amd,df])
```

```python
amd.head()
amd.tail()
amd.sample(12)
```

```python
amd['Order Date']=pd.to_datetime(amd['Order Date'],errors='coerce')
```

```python
amd.info()
```

```python
amd['Month']=amd['Order Date'].dt.month
```

```python
amd.sample(22)
```

```python
amd=amd.dropna()
```

```python
amd['Month']=amd['Month'].astype('int32')
amd['Quantity Ordered']=amd['Quantity Ordered'].astype('int32')
amd['Price Each']=amd['Price Each'].astype('float32')
```

```python
amd.info()
```

```python
amd['Sales']=amd['Price Each']*amd['Quantity Ordered']
```

```python
amd.info()
```

```python
spm=amd.groupby('Month')['Sales'].sum()
spm
```

```python
import matplotlib.pyplot as plt
```

```python
month=range(1,13)
```

```python
plt.bar(month,spm)
plt.xlabel('Months')
plt.ylabel('Sales($)')
plt.xticks(month)
```

```python
amd['City']=amd['Purchase Address'].apply(lambda y: f"{y.split(',')[1]} ({y.split(',')[2].split(' ')[1]})")
```

```python
amd.head()
```

```python
result=amd.groupby('City')['Sales'].sum()
result
```

```python
cities=[city for city,df in amd.groupby('City')]
cities
```

```python
plt.bar(cities,result)
plt.xticks(cities,rotation='vertical')
plt.xlabel('Cities')
plt.ylabel('Sales($)')
```

```python
df=amd[amd['Order ID'].duplicated(keep=False)]
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
```

```python
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
```

```python
df
```

```python
from itertools import combinations
from collections import Counter
```

```python
count =Counter()
```

```python
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
```

```python
count
```

```python
count.most_common(10)
```
